```python
import numpy as np
A = np.array([[1, -2, 3], [4,  0, -1], [-2, 1, 5]])
num_samples = 500000
approx_norm_1 = 0
for _ in range(num_samples):
    v = np.random.randn(A.shape[1])  
    v /= np.linalg.norm(v, 1)  
    norm_ratio = np.linalg.norm(A @ v, 1)  
    approx_norm_1 = max(approx_norm_1, norm_ratio) 

print(f"Approximate Norm 1: {approx_norm_1}")
```

```python
import numpy as np
A = np.array([[1, -2, 3],[4,  0, -1],[-2, 1, 5]])
norm_1 = np.linalg.norm(A, 1)
norm_inf = np.linalg.norm(A, np.inf) 
norm_fro = np.linalg.norm(A, 'fro')  
norm_2 = np.linalg.norm(A, 2)       
print(f"Norm 1 (Column Norm): {norm_1}")
print(f"Infinity Norm (Row Norm): {norm_inf}")
print(f"Frobenius Norm: {norm_fro}")
print(f"Spectral Norm (Largest Singular Value): {norm_2}")
```

```python
import numpy as np
A = np.array([[1, -2, 3], [4,  0, -1], [-2, 1, 5]])
num_samples = 100000
approx_norm_infty = 0
for _ in range(num_samples):
    v = np.random.randn(A.shape[1])  
    v /= np.linalg.norm(v, np.inf)  
    norm_ratio = np.linalg.norm(A @ v, np.inf)  
    approx_norm_infty = max(approx_norm_infty, norm_ratio) 
print(f"Approximate Norm infty: {approx_norm_infty}")
```
